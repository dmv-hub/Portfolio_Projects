```python
#importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #configuration size for plots to be created
```

```python
#reading data

df = pd.read_csv('/Users/eldavopr/Desktop/movies.csv')
```

```python
#looking at data

df.head()
```

```python
#identifying nulls per column

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types per column

df.dtypes
```

```python
df.dropna()
```

```python
# order data by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows', 30)
```

```python
# dropping duplicates

df.drop_duplicates()
```

```python
# correlation hypothesis
# the more budget the higher the gross revenue
df.dtypes
```

```python
df.head()
```

```python
#Scatter plot comparing budget v gross revenue
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "black"})
```

```python
#looking at correlation
#high correlation between budget and gross revenue
df.corr()
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Values')
plt.xlabel('Movie Features')
plt.show()
```

```python
#Turning str to categorical values for correlation

df_to_num = df

for col_name in df_to_num.columns:
    if(df_to_num[col_name].dtype == 'object'):
        df_to_num[col_name] = df_to_num[col_name].astype('category')
        df_to_num[col_name] = df_to_num[col_name].cat.codes
        
df_to_num
        
```

```python
df
```

```python
correlation_matrix = df_to_num.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Values')
plt.xlabel('Movie Features')
plt.show()
```

```python
df_to_num.corr()
```

```python
corr_mat = df_to_num.corr()

corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > .5]
high_corr
```
